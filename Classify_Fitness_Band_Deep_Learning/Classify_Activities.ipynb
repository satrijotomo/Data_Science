## Load Data

```python
import pandas as pd
import numpy as np

test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv")
```

```python
train.head()
```

```python
print('Train Data shape:', train.shape)
print('Test Data shape: ', test.shape)
```

```python
train.Activity.value_counts()
```

```python
test.Activity.value_counts()
```

Data is equally distributed and we don't see any skewness.

```python
# suffling data

from sklearn.utils import shuffle

test = shuffle(test)
train = shuffle(train)
```

```python
# separating data inputs and outputs labels

testData = test.drop(['Activity', 'subject'], axis = 1).values
testLabel = test.Activity.values

trainData = train.drop(['Activity', 'subject'], axis = 1).values
trainLabel = train.Activity.values
```

```python
# encoding labels: create encoder, fit data into encoder, and then transform it

from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()

# encoding test labels
encoder.fit(testLabel)
testLabelE = encoder.transform(testLabel)

#encoding train labels
encoder.fit(trainLabel)
trainLabelE = encoder.transform(trainLabel)
```

```python
# Neural network
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils.np_utils import to_categorical
```

```python
# We use scaler  model is fast,since values are converted to binary rather 
# than the continuous variables
# PCA can be used when there are high degree of linearities between variables

n_input = trainDataSc.shape[1] # number of features
n_output = 6 # number of possible output (Activity)
n_samples = trainDataSc.shape[1] # number of training samples
n_hidden_units = 40
Y_train = to_categorical(trainLabelE) 
Y_test = to_categorical(testLabelE)
print(Y_train.shape), print(Y_test.shape)
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(n_hidden_units, input_dim=n_input, activation='relu'))
    model.add(Dense(n_hidden_units, input_dim=n_input, activation='softmax'))
    #compile model
    model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```
