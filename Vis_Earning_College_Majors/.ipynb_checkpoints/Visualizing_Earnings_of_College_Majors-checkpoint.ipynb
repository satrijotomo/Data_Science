# Visualizing Earnings Based on College Majors

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo (https://github.com/fivethirtyeight/data/tree/master/college-majors)

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.


Questions we would like to answer with the help of visualization:
- Do students in more popular majors make more money?
- How many majors are predominantly male? Predominantly female?
- Which category of majors have the most students?


```python
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
# Drop row with missing data
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
cleaned_data_count
```

## Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample_size vs. Median', figsize=(10,8))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs. Unemployment_rate', figsize=(10,8))
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Median', figsize=(10,8))
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment_rate', figsize=(10,8))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median', figsize=(10,8), xlim=[0,300000])
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median', figsize=(10,8),xlim=[0,300000])
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs. Median', figsize=(10,8))
```

### Finding So Far
- Do students in more popular majors make more money?

Majors with big sample size, hence more popular, does not translate to making more money. There does not seem a strong correlation between popularity and salary.

- Do students that majored in subjects that were majority female make more money?

Last data visualization indicates that students majoring in subjects with female majority receive less salary.  

- Is there any link between the number of full-time employees and median salary?

Again, the dataset shows that there is no strong correlation between full-time employees and median salary.
