```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("factbook.db")
df = pd.read_sql_query("select * from sqlite_master where type='table';", conn)
df
```

```python
df = pd.read_sql_query("select * from facts limit 5;", conn)
df
```

```python
df = pd.read_sql_query("select MIN(population), MAX(population), MIN(population_growth), MAX(population_growth) from facts;",conn)
df
```

### Outliers

```python
query="select * from facts where population=0;"
df = pd.read_sql_query(query,conn)
df
```
