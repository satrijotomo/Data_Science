This project try to find the best model to predict the activity based on the predictor values in the dataset. For each row the activity can be either 'STANDING', 'LAYING','WALKING', 'WALKING_DOWNSTAIRS'or 'WALKING_UPSTAIRS.

### 1. Logistic Regression

```python
## Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C=10, solver='sag')
```

- PCA score: 0.873

Transformation results do not show improvement from the original result.

### 2. Neural network

```python
# Neural network
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils.np_utils import to_categorical
```

```python
# Replace activities with index
replacer = {}
for idx, activity in enumerate(train.Activity.unique()):
    replacer[activity] = idx
print(replacer)
train.Activity = train.Activity.replace(replacer)
test.Activity = test.Activity.replace(replacer)
train.head()
```

```python
train_df = train.drop("subject", axis=1)
test_df = test.drop("subject", axis=1)
```

```python
def get_all_data():
    train_values = train_df.values
    test_values = test_df.values
    np.random.shuffle(train_values)
    np.random.shuffle(test_values)
    X_train = train_values[:, :-1]
    X_test = test_values[:, :-1]
    y_train = train_values[:, -1]
    y_test = test_values[:, -1]
    return X_train, X_test, y_train, y_test
```

```python
# We use scaler  model is fast,since values are converted to binary rather 
# than the continuous variables
# PCA can be used when there are high degree of linearities between variables

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train, X_test, y_train, y_test = get_all_data()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

n_input = X_train.shape[1] # number of features
n_output = 6 # number of possible labels
n_samples = X_train.shape[0] # number of training samples
n_hidden_units = 40
Y_train = to_categorical(y_train)
Y_test = to_categorical(y_test)
print(Y_train.shape)
print(Y_test.shape)

```

```python
def create_model():
    model = Sequential()
    model.add(Dense(n_hidden_units,
                    input_dim=n_input,
                    activation="relu"))
    model.add(Dense(n_hidden_units,
                    input_dim=n_input,
                    activation="relu"))
    model.add(Dense(n_output, activation="softmax"))
    # Compile model
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
    return model
```

```python
estimator = KerasClassifier(build_fn=create_model, epochs=20, batch_size=10, verbose=False)
estimator.fit(X_train, Y_train)
print("Neural Network Score: {}".format(estimator.score(X_test, Y_test)))
```

### 3. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier


X_train, X_test, y_train, y_test = get_all_data()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

model = RandomForestClassifier(n_estimators=500)
model.fit(X_train, y_train)

print("Random Forest Score: {}".format(model.score(X_test, y_test)))
```

### 4. K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=24)
model.fit(X_train, y_train)

print("K-Nearest Neighbors Score: {}".format(model.score(X_test, y_test)))
```

### 5. Support Vector Machine

```python
import time
t0 = time.clock()
from sklearn.svm import SVC
from sklearn.feature_selection import RFECV
from sklearn.model_selection import StratifiedKFold

svc = SVC(kernel="linear")
rfecv = RFECV(estimator=svc, step=1, cv=StratifiedKFold(6),
              scoring='accuracy')
rfecv.fit(X_train, y_train)
print('Support Vector Machine score: {}'.format(rfecv.score(X_test,y_test)))

print('Ranking of features starting from the best estimated \n',rfecv.ranking_)
print('Procesing time in second: ', time.clock() - t0)
```

### 6. Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier

dtClf= DecisionTreeClassifier(criterion='entropy')
tree = dtClf.fit(X_train, y_train)

print('Decision Tree score: {}'.format(dtClf.score(X_test,y_test)))
```
